# Database

Put all your results in an SQLite3 database (remember, SQLite stores its databases as files in your local machine - make sure to create your database in your project's data/ directory!)

```python
def create_connection(path):
    connection = None
    try:
        connection = sqlite3.connect(path)
        print("Connection to SQLite DB successful")
    except Error as e:
        print(f"The error '{e}' occurred")

    return connection

def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query executed successfully")
    except Error as e:
        print(f"The error '{e}' occurred")
        
connection = create_connection("../data/cityBikesPOI.sqlite")

#note that numSlotsTot is not always exactly equal to numOfEmptySlots + numOfBikesAvails for some reason
create_city_bikes_table = """
CREATE TABLE IF NOT EXISTS city_bikes (
    id INTEGER PRIMARY KEY NOT NULL,
    name TEXT,
    numOfEmptySlots INTEGER,
    numOfBikesAvails INTEGER,
    numSlotsTot INTEGER,
    latitude REAL,
    longitude REAL
);
"""
create_yelp_POIall_table = """
CREATE TABLE IF NOT EXISTS yelp_POIall (
    bikeStationID INTEGER NOT NULL,
    name TEXT,
    latitude REAL,
    longitude REAL,
    category TEXT,
    rating REAL,
    numReviews INTEGER,
    FOREIGN KEY (bikeStationID) REFERENCES city_bikes (id) 
);
"""
create_yelp_restaurants_table = """
CREATE TABLE IF NOT EXISTS yelp_restaurants (
    bikeStationID INTEGER NOT NULL,    
    name TEXT,
    latitude REAL,
    longitude REAL,
    category TEXT,
    rating REAL,
    numReviews INTEGER,
    FOREIGN KEY (bikeStationID) REFERENCES city_bikes (id)   
);
"""

create_yelp_coffee_table = """
CREATE TABLE IF NOT EXISTS yelp_coffee (
    bikeStationID INTEGER NOT NULL,
    name TEXT,
    latitude REAL,
    longitude REAL,
    category TEXT,
    rating REAL,
    numReviews INTEGER,
    FOREIGN KEY (bikeStationID) REFERENCES city_bikes (id)  
);
"""

create_fsV2_POIall_table = """
CREATE TABLE IF NOT EXISTS fsV2_POIall (
    bikeStationID INTEGER NOT NULL,
    name TEXT,
    latitude REAL,
    longitude REAL,
    category TEXT,
    FOREIGN KEY (bikeStationID) REFERENCES city_bikes (id)  
);
"""

create_fs_dine_drink_table = """
CREATE TABLE IF NOT EXISTS fs_dine_drink (
    bikeStationID INTEGER NOT NULL,
    name TEXT,
    latitude REAL,
    longitude REAL,
    category TEXT,
    FOREIGN KEY (bikeStationID) REFERENCES city_bikes (id)    
);
"""

execute_query(connection, create_city_bikes_table)
execute_query(connection, create_yelp_POIall_table)
execute_query(connection, create_yelp_restaurants_table)
execute_query(connection, create_yelp_coffee_table)
execute_query(connection, create_fsV2_POIall_table)
execute_query(connection, create_fs_dine_drink_table)
```

```python
df_cb.to_sql('city_bikes',connection,if_exists='replace',index=False)
connection.commit()
df_fs_dine_drink.to_sql('yelp_POIall',connection,if_exists='replace',index=False)
connection.commit()
df_fsV2all.to_sql('yelp_restaurants',connection,if_exists='replace',index=False)
connection.commit()
df_yelp_restaurants.to_sql('yelp_coffee',connection,if_exists='replace',index=False)
connection.commit()
df_yelp_coffee.to_sql('fsV2_POIall',connection,if_exists='replace',index=False)
connection.commit()
df_yelp_POIall.to_sql('fs_dine_drink',connection,if_exists='replace',index=False)
connection.commit()

```
